#### 2: Write a user-defined function for origin  


```python
#use apply function
#1 -USA ,2-EUROPE ,3 - ASIA
def origin(num):
  if num==1:
    return 'USA'
  elif num==2:
    return 'EUROPE'
  else:
    return 'ASIA'

Data_set['origin']=Data_set['origin'].apply(origin)



```

#### 3: Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4


```python
#Drawing the graph
sns.pairplot(Data_set[['mpg','acceleration','origin']],hue='origin', size=4)
```
